### Existe-t-il une diff√©rence significative dans la valeur m√©diane des maisons bord√©es par la rivi√®re Charles ou non? (Test T pour les √©chantillons ind√©pendants)

```python
print("H0 : There is no statistically significant difference in the median value of homes on or off the Charles River.")
print("H1 : There is a statistically significant difference in the median value of homes on or off the Charles River.")

#Tester la variance avec un test de Levene
levene_statistic, levene_pvalue = sp.stats.levene(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                                                  boston_df[boston_df['CHAS'] == 1]['MEDV'], center = 'mean')

#Faire le t test en ad√©quation avec le r√©sultat pr√©c√©dent
t_statistic, t_pvalue = sp.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                                           boston_df[boston_df['CHAS'] == 1]['MEDV'], equal_var = False)

print('')
print("Levene Test p-value :", levene_pvalue)
print("Welch T Test p-value :", t_pvalue)
print('')
print("Since the p-value of the T test is less than 0.05, it is possible to reject the null hypothesis (HO) and \nsuggest that houses on the banks of the Charles River are more expensive than houses that aren't.")
```

### Y a-t-il une diff√©rence dans les valeurs m√©dianes des maisons (MEDV) pour chaque proportion d'unit√©s occup√©es par le propri√©taire et construites avant 1940 (AGE)? (ANOVA)

```python
print("H0 : There is no statistically significant difference in median home values by the proportion of owner-occupied units built before 1940")
print("H1 : There is a statistically significant difference in median home values by the proportion of owner-occupied units built before 1940")

# Cr√©ation des groupes

# Une premi√®re cat√©gorie sera les quartiers qui ont de 0% √† 25% de logements construits avant 1940 et qui sont habit√© par leur propri√©taire
boston_df.loc[(boston_df['AGE'] >= 0) & (boston_df['AGE'] <= 25), 'age_group'] = '0% to 25%'

# Une deuxi√®me cat√©gorie sera les quartiers qui ont de 25 √† 50% de logements construits avant 1940 et qui sont habit√© par leur propri√©taire
boston_df.loc[(boston_df['AGE'] > 25) & (boston_df['AGE'] <= 50), 'age_group'] = '25% to 50%'

# Une troisi√®me cat√©gorie sera les quartiers qui ont de 50 √† 75% de logements construits avant 1940 et qui sont habit√© par leur propri√©taire
boston_df.loc[(boston_df['AGE'] > 50) & (boston_df['AGE'] <= 75), 'age_group'] = '50% to 75%'

# Une quatri√®me cat√©gorie sera les quartiers qui ont de 75% √† 100% de logements construits avant 1940 et qui sont habit√© par leur propri√©taire
boston_df.loc[(boston_df['AGE'] > 75) & (boston_df['AGE'] <= 100), 'age_group'] = '75% to 100%'


# Test de Levene pour voir s'il y a une diff√©rence statistiquement significative des variances.
levene_statistic, levene_pvalue = sp.stats.levene(boston_df[boston_df['age_group'] == '0% to 25%']['MEDV'],
                                                  boston_df[boston_df['age_group'] == '25% to 50%']['MEDV'],
                                                  boston_df[boston_df['age_group'] == '50% to 75%']['MEDV'],
                                                  boston_df[boston_df['age_group'] == '75% to 100%']['MEDV'],
                                                  center = 'mean')

# S√©parer les groupes (un pour chaque cat√©gorie) en variable unique
twentyfive_lower   = boston_df[boston_df['age_group'] == '0% to 25%']['MEDV']
twentyfive_fifty   = boston_df[boston_df['age_group'] == '25% to 50%']['MEDV']
fifty_seventyfive  = boston_df[boston_df['age_group'] == '50% to 75%']['MEDV']
seventyfive_hundred= boston_df[boston_df['age_group'] == '75% to 100%']['MEDV']

# Test Anova
f_statistic, f_pvalue = sp.stats.f_oneway(twentyfive_lower, twentyfive_fifty, fifty_seventyfive, seventyfive_hundred)

print('')
print("Levene Test p-value :", levene_pvalue)
print("Anova Test p-value :", f_pvalue)
print('')
print("Since the p-value of the Anova test is less than 0.05, it is possible to reject the null hypothesis (H0) and suggest that \nthere is a statistically significant difference in the median house values depending on the proportion of owner-occupied \nunits built before 1940.")
```

### Peut-on conclure qu'il n'y a pas de relation entre les concentrations d'oxyde nitrique et la proportion d'hectares de commerces autres que de d√©tail par ville (corr√©lation de Pearson) ?

```python
print("H0 : There is no statistically significant relationship between nitric oxide concentrations and the proportion of non-retail hectares by city")
print("H1 : There is a statistically significant relationship between nitric oxide concentrations and the proportion of non-retail hectares by city")
print('')

# Calculer le graphique
ax = sns.scatterplot(x='INDUS', y='NOX', data=boston_df, s=50, color='green')
plt.show()

# Test de corr√©lation
statistic, pvalue = sp.stats.pearsonr(boston_df['INDUS'], boston_df['NOX'])

print('')
print("Pearson Test p-value :", pvalue)
print('')
print("Since the p-value of the Pearson test is less than 0.05, it is possible to reject the null hypothesis (H0) and\nsuggest that there is a statistically significant relationship between nitric oxide concentrations and the proportion\nhectares of non-retail businesses per city.")
```

### Quel est l'impact d'une distance pond√©r√©e suppl√©mentaire par rapport aux cinq centres d'emploi de Boston sur la valeur m√©diane des maisons occup√©es par leur propri√©taire ? (Analyse de r√©gression)

```python
print("H0 : There is no relationship between the weighted distance to five Boston employment centers and the median value of owner-occupied homes.")
print("H1 : There is a relationship between the weighted distance to five Boston employment centers and the median value of owner-occupied homes.")
print("")

ax = sns.scatterplot(x='DIS', y='MEDV', data=boston_df, s=50, color='red')
plt.show()

# Identification de la variable ind√©pendante
X = boston_df['DIS']

# Identification de la variable d√©pendante
y = boston_df['MEDV']

# Ajouter une constante au mod√®le
X = sm.add_constant(X)

# Calculer le mod√®le
model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Pr√©sentation des r√©sultats
print(model.summary())

print('')
print("Since the p-value of the regression analysis is less than 0.05, it is possible to reject the null hypothesis (H0) \nand suggest that there is a statistically significant relationship between the weighted distance from five Boston \nemployment centers and the median value of owner-occupied homes.")
print('')
print("For a one unit increase in the DIS variable, the MEDV variable will increase on average by 1.09 units.")
print('')
print("In other words, an increase of one mile in the weighted distance from the \nfive employment centers in Boston will, on average, result in a $1,090 \nincrease in the median value of owner-occupied homes.")
```
